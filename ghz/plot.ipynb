```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
for N in [4, 6, 8, 10]:
    plt.figure()
    plt.yscale('log')
    for lr in [0.1, 0.05, 0.01]:
        plt.plot(10 + np.arange(50) * 10, np.load(f'results/GHZ_N{N}lr{lr}.npz'),
                label=f'lr={lr}')
    plt.legend()
    plt.title(f'N={N}')
    plt.show()
```
